```python
import numpy as np
import pandas as pd

# Generating  temperature data
np.random.seed(42)
temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))

# Creating city names and months
cities = [f"City_{i}" for i in range(1, 11)]
months = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

#Converting to Pandas DataFrame
df_temperatures = pd.DataFrame(data=temperature_data, index=cities, columns=months)


print("Temperature DataFrame:")
print(df_temperatures.head())
```

```python
#Calculating annual average temperature for each city
annual_avg_temp = df_temperatures.mean(axis=1)

#Identifing cities with the highest and lowest average temperatures
city_highest_temp = annual_avg_temp.idxmax()
city_lowest_temp = annual_avg_temp.idxmin()

print("\nAnnual Average Temperatures:")
print(annual_avg_temp)

print(f"\nCity with the Highest Average Temperature: {city_highest_temp}")
print(f"City with the Lowest Average Temperature: {city_lowest_temp}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
for city in df_temperatures.index:
    plt.plot(months, df_temperatures.loc[city], label=city)
plt.title('Monthly Temperature Trends for Cities')
plt.xlabel('Months')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
annual_avg_temp.sort_values().plot(kind='bar', color='orange')
plt.title('Annual Average Temperature by City')
plt.xlabel('City')
plt.ylabel('Average Temperature (°C)')
plt.grid(axis='y')
plt.show()
```
