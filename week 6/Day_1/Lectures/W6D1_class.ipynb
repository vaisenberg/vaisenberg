```python
sales_data = pd.DataFrame({
    "Monday": [200, 150, 100, 300, 50],
    "Tuesday": [220, 160, 110, 320, 60],
    "Wednesday": [250, 180, 120, 340, 70],
    "Thursday": [270, 200, 130, 360, 80],
    "Friday": [300, 210, 140, 400, 90],
    "Saturday": [310, 220, 150, 420, 100],
    "Sunday": [280, 190, 160, 410, 110]
}, index=["Store 1", "Store 2", "Store 3", "Store 4", "Store 5"])
print("Sales Data (Pandas DataFrame):\n", sales_data)
```

```python
sales_array = sales_data.to_numpy()
np.save('sales_data.npy', sales_array)
print("Sales Data (NumPy Array):\n", sales_array)
```

```python

daily_sales = sales_array.sum(axis=0)

# Step 3: Find the days where the total sales are greater than 1200
days_above_1200 = daily_sales[daily_sales > 1200]

# Step 4: Display the result
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
result_days = [days[i] for i in range(len(daily_sales)) if daily_sales[i] > 1200]

print("Days with sales greater than 1200:", result_days)
```
