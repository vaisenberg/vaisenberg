data = [12, 15, 13, 12, 18, 20, 22, 21]
Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.


```python
from scipy.stats import describe

data = [12, 15, 13, 12, 18, 20, 22, 21]
stats = describe(data)
mean = stats.mean
variance = stats.variance
std_dev = variance**0.5  # Standard deviation is the square root of variance
median = sorted(data)[len(data)//2] if len(data) % 2 == 0 else sorted(data)[len(data)//2] #The middle value of the dataset when sorted.

mean, median, variance, std_dev
```

Exercise 3
Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.

SciPy's scipy.stats.norm module provides a rvs method to generate random samples from a normal distribution.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

mean = 50
std_dev = 10

# Generating random samples
random_data = norm.rvs(loc=mean, scale=std_dev, size=10000)

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(random_data, bins=30, density=True, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Normal Distribution Using SciPy', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

The plot displays the normal distribution with a mean of 50 and a standard deviation of 10. The bell-shaped curve represents the probability density function (PDF), showing the likelihood of values occurring within this distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 50
std_dev = 10

# Generating data points for the x-axis
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Generating the probability density function (PDF) for the normal distribution
pdf = norm.pdf(x, mean, std_dev)

# Plotting the normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Normal Distribution (mean={mean}, std_dev={std_dev})', color='blue')
plt.title('Normal Distribution', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```

Exercise 4 : T-Test Application

```python
from scipy.stats import ttest_ind
import numpy as np

# Generating two sets of random data
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, StdDev=10, Sample Size=1000
data2 = np.random.normal(loc=35, scale=15, size=1000)  # Mean=35, StdDev=15, Sample Size=1000

# Performing a two-sample (independent) T-test
t_stat, p_value = ttest_ind(data1, data2)


t_stat, p_value
```

t-stat: This large positive value wchich is greater than 24  indicates a significant difference between the means of the two datasets, with data1 having a much higher mean than data2.
p-value is is extremely small and suggests that the difference in means is statistically significant
The T-test reveals that the two datasets are significantly different in their means
