Reminder: Eigenvectors are used to understand the linear transformations of a matrix. They show the directions in which a matrix, when applied, stretches the space.

Matrix normalization is the process of adjusting the values in the matrix to a common scale, without distorting differences in the ranges of values. It’s used to bring all the variables to a uniform scale. This is particularly useful in machine learning algorithms that calculate distances between data.

Create a 5x5 matrix with random numbers and perform the following operations:
1. Calculate the eigenvalues and eigenvectors.
2. Perform matrix normalization.

```python
import numpy as np

# Create a 5x5 matrix with random numbers
np.random.seed(42)  # For reproducibility
matrix = np.random.randint(1, 100, size=(5, 5))

# 1. Calculate the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(matrix)

# 2. Perform matrix normalization
# Normalize each element by dividing by the maximum value in the matrix
normalized_matrix = matrix / np.max(matrix)

# Display results
(matrix, eigenvalues, eigenvectors, normalized_matrix)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generating monthly sales data as a NumPy array
np.random.seed(42)  # For reproducibility
monthly_sales_data = np.random.randint(100, 500, size=12)

# Preparing data for linear regression
months = np.arange(1, 13).reshape(-1, 1)  # Months 1 to 12
sales = monthly_sales_data  # Monthly sales data

# Create and fit the linear regression model
model = LinearRegression()
model.fit(months, sales)

# Extracting the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Forecast future data points (e.g., for the next 6 months)
future_months = np.arange(13, 19).reshape(-1, 1)  # Months 13 to 18
forecasted_sales = model.predict(future_months)

# Visualize the data and forecast
plt.figure(figsize=(10, 6))
plt.scatter(months, sales, color='blue', label='Actual Sales')
plt.plot(months, model.predict(months), color='green', label='Regression Line')
plt.scatter(future_months, forecasted_sales, color='red', label='Forecasted Sales')
plt.plot(future_months, forecasted_sales, color='red', linestyle='--')
plt.title('Monthly Sales and Forecast', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Print slope, intercept, and forecasted sales
print(f"Slope (m): {slope}")
print(f"Intercept (b): {intercept}")
print("Forecasted Sales for Next 6 Months:")
print(forecasted_sales)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Generating another sales dataset (e.g., for comparison)
np.random.seed(24)  # New seed for reproducibility
monthly_sales_data_alt = np.random.randint(100, 500, size=12)

# Model 1: Original sales data
model1 = LinearRegression()
model1.fit(months, sales)
predictions_model1 = model1.predict(months)

# Model 2: Alternative sales data
model2 = LinearRegression()
model2.fit(months, monthly_sales_data_alt)
predictions_model2 = model2.predict(months)

# Evaluate and compare the models
mae_model1 = mean_absolute_error(sales, predictions_model1)
mae_model2 = mean_absolute_error(monthly_sales_data_alt, predictions_model2)

mse_model1 = mean_squared_error(sales, predictions_model1)
mse_model2 = mean_squared_error(monthly_sales_data_alt, predictions_model2)

r2_model1 = model1.score(months, sales)
r2_model2 = model2.score(months, monthly_sales_data_alt)

# Display results
print(f"Model 1 (Original Data):")
print(f"  R-squared: {r2_model1:.2f}, MAE: {mae_model1:.2f}, MSE: {mse_model1:.2f}")

print(f"\nModel 2 (Alternative Data):")
print(f"  R-squared: {r2_model2:.2f}, MAE: {mae_model2:.2f}, MSE: {mse_model2:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Generate a dataset following a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, StdDev=10, Size=1000

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Normally Distributed Data', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Calculate skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data)

(data_skewness, data_kurtosis)
```

```python
import pandas as pd

# Creating a sample Pandas DataFrame
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product A', 'Product B'],
    'Sales': [200, 150, 300, 250, 180],
    'Month': ['January', 'February', 'March', 'April', 'May']
}
df = pd.DataFrame(data)

# Grouping and aggregation
# Summarizing total sales by product
total_sales_by_product = df.groupby('Product')['Sales'].sum()

# Calculating average sales by product
average_sales_by_product = df.groupby('Product')['Sales'].mean()

# Grouping by product and aggregating multiple statistics
summary_stats_by_product = df.groupby('Product')['Sales'].agg(['sum', 'mean', 'max', 'min'])



total_sales_by_product, average_sales_by_product, summary_stats_by_product
```
