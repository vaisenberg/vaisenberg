**Exercise 1 : Linear Regression Analysis**

The slope and interecept is ration of two variables (by simple words). The slope indicates the steepness of a line and the intercept indicates the location where it intersects an axis. The slope and the intercept define the linear relationship between two variables, and can be used to estimate an average rate of change. The greater the magnitude of the slope, the steeper the line and the greater the rate of change.

But while predicting the price of the house The intercept just doesn’t have any meaningful interpretation for this model.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

house_sizes = np.array([50, 70, 80, 100, 120]).reshape(-1, 1)
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

model = LinearRegression()
model.fit(house_sizes, house_prices)

slope = model.coef_[0]
intercept = model.intercept_

slope, intercept
```

Assuming that there are no other constrains on the housing prices , it is possible to apply sklearn.linear_model assuming the price of a house that is 90 square meters is a linear combination of the input house_seizes and house_prices.



```python
new_house_size = 90
predicted_price_90 = model.predict(np.array([[new_house_size]]))[0]

predicted_price_90
```

the plot for the linear regression

```python
import matplotlib.pyplot as plt

# Plot the data points
plt.scatter(house_sizes, house_prices, color='blue', label='Data points')

# Plot the regression line
regression_line = slope * house_sizes + intercept
plt.plot(house_sizes, regression_line, color='red', label='Regression line')

# Labels and title
plt.title('House Size vs. Price')
plt.xlabel('House Size (square meters)')
plt.ylabel('House Price (currency units)')
plt.legend()

# Display the plot
plt.show()
```

```python
predicted_point_size = 90
predicted_point_price = predicted_price_90

plt.scatter(house_sizes, house_prices, color='blue', label='Data points')
plt.plot(house_sizes, regression_line, color='red', label='Regression line')
plt.scatter(predicted_point_size, predicted_point_price, color='green', label='Predicted point (90 sqm)', zorder=5)
plt.title('House Size vs. Price with Prediction')
plt.xlabel('House Size (square meters)')
plt.ylabel('House Price (currency units)')
plt.legend()

plt.show()

```

**Exercise 2 : Understanding ANOVA**

Analysis of Variance (ANOVA) is a statistical technique used to compare means of three samples of plants to wchich defferent types of fertilozer have been applied. It helps in determining whether there are any statistically significant differences between the means of 3 gtoups

```python
from scipy.stats import f_oneway
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
f_value, p_value = f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("F-value:", f_value)
print("P-value:", p_value)
```

F-value 28.53 indicates that there is a substantial variance between groups compared to within groups.
P-value 0.000028, since the P-value is much smaller than 0.05, we reject the null hypothesis. This means the fertilizers have significantly different effects on plant growth.

**Exercise 3 : Probability Distributions**

```python
from scipy.stats import binom

n = 10  # Number of trials
k = 5   # Number of successes
p = 0.5 # Probability of success

# Calculate probability
probability = binom.pmf(k, n, p)

print(f"The probability of getting exactly {k} heads in {n} coin flips is: {probability}")
```

**Exercise 4 : Correlation Coefficients**

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

pearson_corr, pearson_p_value = pearsonr(data['age'], data['income'])
spearman_corr, spearman_p_value = spearmanr(data['age'], data['income'])

print("Pearson Correlation Coefficient:", pearson_corr)
print("Spearman Correlation Coefficient:", spearman_corr)
```

Create a scatter plot of age vs. income

adding regression line

```python
plt.scatter(data['age'], data['income'], color='blue', label='Data Points', alpha=0.7)
m, b = np.polyfit(data['age'], data['income'], 1)  # Linear regression
plt.plot(data['age'], m * data['age'] + b, color='red', label='Regression Line', linestyle='--')
plt.grid(alpha=0.3)
plt.title('Enhanced Scatter Plot of Age vs. Income with Regression Line')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()


plt.show()
```
