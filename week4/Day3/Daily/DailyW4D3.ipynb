```python
# Load an image using PIL
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
rotated_image = original_image.rotate(90)
plt.imshow(rotated_image)
```

```python
flipped_horizontally = ImageOps.mirror(original_image)
plt.imshow(flipped_horizontally)
```

```python
flipped_vertically = ImageOps.flip(original_image)
plt.imshow(flipped_vertically)
```

```python
width, height = original_image.size


new_size = (int(width * 1.2), int(height * 1.2))

zoomed_image = original_image.resize(new_size)


zoomed_image.show()
```

```python
import matplotlib.pyplot as plt
plt.imshow(zoomed_image)
plt.axis('off')
plt.show()
```
