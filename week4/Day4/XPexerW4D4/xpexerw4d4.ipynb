```python
weatherdata_munich.columns = weatherdata_munich.columns.str.strip().str.lower()
weatherdata_munich.info()






```

```python
unique_counts_columns = weatherdata_munich.nunique()
print(unique_counts_columns)
```

```python
import numpy as np
import numpy as np


weatherdata_munich['mess_datum'] = pd.to_datetime(weatherdata_munich['mess_datum'], format='%Y%m%d', errors='coerce')
weatherdata_munich.info()
weatherdata_munich.head()
print(weatherdata_munich.duplicated())
weatherdata_munich.describe()

```

```python
weatherdata_munich_formated = weatherdata_munich.drop(columns=['eor'])

# Check the structure of the new DataFrame
weatherdata_munich_formated.info()
```

```python
weatherdata_munich_formated['Year'] = weatherdata_munich_formated['mess_datum'].dt.year
weatherdata_munich_formated['Month'] = weatherdata_munich_formated['mess_datum'].dt.month
weatherdata_munich_formated['Week'] = weatherdata_munich_formated['mess_datum'].dt.isocalendar().week
weatherdata_munich_formated.head()
weatherdata_munich_formated.describe()
```

```python
 print(weatherdata_munich_formated[['Year', 'Month', 'Week']].head())
print(weatherdata_munich_formated.dtypes)




```

```python
weatherdata_munich_formated_filled = weatherdata_munich_formated.groupby(['Year', 'Week']).apply(
    lambda group: group.fillna(group.mean(numeric_only=True))
)
weatherdata_munich_formated_filled.info()
```

```python
weatherdata_munich_formated_filled = weatherdata_munich_formated.replace(-999, np.nan)


weatherdata_munich_formated_filled = weatherdata_munich_formated_filled.groupby(['Year', 'Week']).apply(
    lambda group: group.fillna(group.mean(numeric_only=True))
)


weatherdata_munich_formated_filled.reset_index(drop=True, inplace=True)


weatherdata_munich_formated_filled.info()
```

```python
weatherdata_munich_formated_filled_reset = weatherdata_munich_formated_filled.reset_index(drop=True)
weatherdata_munich_formated_filled_reset.info()
weatherdata_munich_formated_filled_reset.describe()
```

```python
completely_mising_months = weatherdata_munich_formated_filled_reset.groupby(['Year', 'Month']).apply(
    lambda group: group.isnull().all()
)
print(completely_mising_months)
```

```python
years_mising_data = weatherdata_munich_formated_filled_reset.groupby(['Year']).apply(
    lambda group: group.isnull().all()
)
missing_years = years_mising_data[years_mising_data.all(axis=1)]
print(missing_years)
missing_years.info()
```

```python
weatherdata_munich_formated_filled.drop(columns=['fx','fm','qn_3','vpm', 'pm','tgk','upm','sdk','nm'], inplace=True)
weatherdata_munich_formated_filled.info()
```

```python
missing_shk_tag = weatherdata_munich_formated_filled[weatherdata_munich_formated_filled['shk_tag'].isnull()]
print(missing_shk_tag)
shk_tag_mode = weatherdata_munich_formated_filled['shk_tag'].mode()[0]
print(shk_tag_mode)
weatherdata_munich_formated_filled['shk_tag'].fillna(shk_tag_mode, inplace=True)
print(weatherdata_munich_formated_filled.isnull().sum())
```

```python
weatherdata_munich_formated_filled.to_csv('weatherdata_munich_formated_filled.csv', index=False)
weatherdata_munich_formated_filled.info()
weatherdata_munich_formated_filled.describe()
weatherdata_munich_formated_filled.head()
```
