```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
sleep_data_US = pd.read_csv('Time Americans Spend Sleeping.csv')
sleep_data_US.head()
```

```python
interesting_columns = ['Year', 'Avg hrs per day sleeping','Standard Error', 'Age Group']
sleep_data_US[interesting_columns].head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.plot(sleep_data_US['Year'], sleep_data_US['Avg hrs per day sleeping'], marker='o', color='blue', label='Average Hours')
plt.xlabel('Year')
plt.ylabel('Average Hours')
plt.title('Average Hours of Sleep Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.errorbar(
    sleep_data_US['Year'],
    sleep_data_US['Avg hrs per day sleeping'],
    yerr=sleep_data_US['Standard Error'],
    fmt='-o',
    color='blue',
    ecolor='red',
    capsize=5,
    label='Average Hours with Error')
plt.xlabel('Year')
plt.ylabel('Average Hours')
plt.title('Average Hours of Sleep Over Time with Error Bars')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(
    sleep_data_US['Year'],
    sleep_data_US['Standard Error'],
    marker='o',
    linestyle='-',
    color='orange',
    label='Standard Error'
)
plt.xlabel('Year')
plt.ylabel('Standard Error')
plt.title('Standard Error of Sleep Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array(sleep_data_US['Year']).reshape(-1, 1)
y = sleep_data_US['Avg hrs per day sleeping']

# Fit the regression model
model = LinearRegression()
model.fit(X, y)

# Predict using the model
sleep_data_US['Predicted Hours'] = model.predict(X)

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(sleep_data_US['Year'], sleep_data_US['Avg hrs per day sleeping'], color='blue', label='Actual Hours')
plt.plot(sleep_data_US['Year'], sleep_data_US['Predicted Hours'], color='red', label='Linear Regression Line')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Avg hrs per day sleeping', fontsize=14)
plt.title('Year vs. Avg hrs per day sleeping', fontsize=16)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

```

```python
age_group_sleep = sleep_data_US.groupby('Age Group')['Avg hrs per day sleeping'].mean()
print(age_group_sleep)
```

```python
age_group_sleep.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Average Hours of Sleep')
plt.title('Average Hours of Sleep by Age Group')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
