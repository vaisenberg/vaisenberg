```python
from google.colab import files
uploaded = files.upload()

iris = pd.read_csv("Iris.csv")
iris.head()

```

```python
iris.dtypes
```

SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm) are of type float64, and the Species column is of type object -string data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

iris['SepalLengthCm'].mean()

```

```python
iris['SepalLengthCm'].median()
```

```python
iris['SepalLengthCm'].mode()[0]
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 6))
plt.hist(iris['SepalLengthCm'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=iris['SepalLengthCm'], color='lightgreen')
plt.title('Box Plot of Sepal Length')
plt.xlabel('Sepal Length (cm)')
```

Mean: The mean sepal length is the average length of the sepal in the dataset.
Median: The median represents the middle value in the sepal length data.
Mode: The mode shows the most frequent value of the sepal length.
Visualizations: Histograms and box plots give insights into the distribution, spread, and presence of any outliers in the sepal length data
