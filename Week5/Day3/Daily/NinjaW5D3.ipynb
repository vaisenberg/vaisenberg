```python
import pandas as pd
import plotly.express as px
from sklearn.datasets import load_iris
iris_data = load_iris()
df = pd.DataFrame(data=iris_data['data'], columns=iris_data['feature_names'])
df['species'] = pd.Categorical.from_codes(iris_data['target'], iris_data['target_names'])
fig = px.scatter(df,
                 x="sepal length (cm)", y="sepal width (cm)",
                 marginal_x="histogram", marginal_y="histogram",
                 title="Scatter Plot of Sepal Length vs Sepal Width",
                 labels={"sepal length (cm)": "Sepal Length (cm)", "sepal width (cm)": "Sepal Width (cm)"},
                 color="species",
                 color_discrete_sequence=px.colors.qualitative.Set1)

fig.show()






```

```python
def plot_type(species):
    if species == 'setosa':
        return 'scatter'
    else:
        return 'line'
df['plot_type'] = df['species'].apply(plot_type)

fig = px.scatter(df,
                 x="sepal length (cm)", y="sepal width (cm)",
                 facet_col="species",
                 color="species",
                 symbol="plot_type",
                 title="Iris Data: Sepal Length vs Sepal Width by Species",
                 labels={"sepal length (cm)": "Sepal Length (cm)", "sepal width (cm)": "Sepal Width (cm)"},
                 hover_data=["species", "sepal length (cm)", "sepal width (cm)"],
                 category_orders={"species": ["setosa", "versicolor", "virginica"]},
                 color_discrete_sequence=px.colors.qualitative.Set1)

fig.update_traces(marker=dict(size=12))
fig.show()
```
